```python
import argparse
import matplotlib.pyplot as plt
```

```python
def extract_info_from_file(log_file):

    result_y = []
    result_x = []
    with open(log_file, 'r') as f:
        for line in f.readlines():
            ss = line.split(' ')
            if ss[0] == 'step:':
                result_x.append(int(ss[1][0:-1])+1)
                result_y.append(float(ss[3][0:-1]))
            elif ss[0] == 'DLL':
                if ss[4] == 'Iteration:':
                    result_x.append(int(ss[5]))
                    result_y.append(float(ss[21]))

    return result_x,result_y
```

```python
tf_x, tf_y = extract_info_from_file('./bert_b32_fp32_1.log')
of_x, of_y = extract_info_from_file('./oneflow_bert_display_loss_only_1.logs')

l1, = plt.plot(tf_x, tf_y, label='tf total loss')
l2, = plt.plot(of_x[0:len(tf_x)], of_y[0:len(tf_y)], color='red', label='of total loss')
plt.legend(loc = 'upper right')
plt.show()
```
